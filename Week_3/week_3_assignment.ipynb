```python
df1 = df.drop("sl_no", axis="columns")
df1.head(5)
```

```python
df1["salary"].fillna(0,inplace=True)
df1.head(5)
```

```python
df1.isnull().sum()
```

```python
# df2 = pd.get_dummies(df1,drop_first=True)
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
df2 = df1
df2["gender"] = encoder.fit_transform(df1["gender"])    # 1 - Male & 0 - Female
df2["ssc_b"] = encoder.fit_transform(df1["ssc_b"])      # 1 - Others & 0 - Central
df2["hsc_b"] = encoder.fit_transform(df1["hsc_b"])      # 1 - Others & 0 - Central
df2["hsc_s"] = encoder.fit_transform(df1["hsc_s"])      
df2["degree_t"] = encoder.fit_transform(df1["degree_t"])
df2["workex"] = encoder.fit_transform(df1["workex"])
df2["specialisation"] = encoder.fit_transform(df1["specialisation"])
df2["status"] = encoder.fit_transform(df1["status"])
```

```python
df2["hsc_s"].unique()
```

```python
corelation_matrix = df2.corr()
corelation_matrix
```

Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sn
```

Gender Distribution in Campus

```python
plt.pie(df["gender"].value_counts(), labels=["Male", "Female"],autopct="%.0f%%")
plt.show()
```

Distribution of salaries for placed students

```python
plt.figure(figsize=(10,7))
sn.histplot(df[df["status"] == "Placed"],x='salary',bins=20, kde=True,color="red")
plt.title("Distribution of salaries for placed students")
plt.show()
```

```python
plt.figure(figsize=(12,10))
sn.heatmap(corelation_matrix,annot=True)
plt.show()
```
